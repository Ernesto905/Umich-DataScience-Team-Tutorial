```python
import pandas as pd
```

```python
# Read in the cereal.csv at ../../data/cereal.csv
```

```python
df = pd.read_csv("../../data/cereal.csv")
```

```python
# Add a serial number column with increasing values, starting at 1, 
# up until the size of our rows
# Then, set the serial number as the index 
```

```python
df["Serial number"] = [i for i in range(1, len(df) + 1)]
df = df.set_index("Serial number")
df
```

```python
# Assume that there are whitespaces in our columns, use the rename method to account for those
# Then, use the rename method to change the columns carbo and potass 
# to carbohydrates and potassium
```

```python
df = df.rename(mapper=str.strip, axis="columns")
df = df.rename(columns={"carbo":"Carbohydrates", "potass":"potassium"})
df
```

```python
# Create a list of the column names
# Then, use a list comprehension to make everything uppercase
# Finally, overwrite the columns with this new list. Do not use an "=" in this final command
```

```python
cols = df.columns
cols = [i.strip().upper() for i in cols]
df.columns=cols
df
```
